```python
colunas_categoricas = ['Migração', 'Sexo','Estrangeiro', 'Necessidades educacionais especiais', 'Devedor',
                     'Taxas de matrícula em dia', 'Bolsista', 'Período','Estado civil', 'Curso', 'Qualificação prévia']
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='if_binary')
df_categorico = df[colunas_categoricas]
data_encoded = pd.DataFrame(encoder.fit_transform(df_categorico).toarray(), columns=encoder.get_feature_names_out(df_categorico.columns))
df_final = pd.concat([df.drop(columns=colunas_categoricas), data_encoded], axis=1)
df_final
```

```python
X = df_final.drop(columns='Target', axis=1)
y = df_final['Target']
```

```python
from sklearn.model_selection import train_test_split
X, X_test, y, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=0)
X_treino, X_val, y_treino, y_val = train_test_split(X, y, stratify=y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_rf = RandomForestClassifier(random_state=0)
modelo_rf.fit(X_treino, y_treino)
y_pred_rf = modelo_rf.predict(X_val)
print('Acurácia de treino: ', modelo_rf.score(X_treino, y_treino))
print('Acurácia de validação: ', modelo_rf.score(X_val, y_val))
print('Acurácia de teste: ', modelo_rf.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo_dt = DecisionTreeClassifier(random_state=0, max_depth=10)
modelo_dt.fit(X_treino, y_treino)
y_pred_dt = modelo_dt.predict(X_val)
print('Acurácia de treino: ', modelo_dt.score(X_treino, y_treino))
print('Acurácia de validação: ', modelo_dt.score(X_val, y_val))
print('Acurácia de teste: ', modelo_dt.score(X_test, y_test))
```

```python
from sklearn.tree import ExtraTreeClassifier

modelo_et = ExtraTreeClassifier(random_state=0, max_depth=10, max_features=1231)
modelo_et.fit(X_treino, y_treino)
y_pred_et = modelo_et.predict(X_val)
print('Acurácia de treino: ', modelo_et.score(X_treino, y_treino))
print('Acurácia de validação: ', modelo_et.score(X_val, y_val))
print('Acurácia de teste: ', modelo_et.score(X_test, y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelo_knc = KNeighborsClassifier(n_neighbors=8)
modelo_knc.fit(X_treino, y_treino)
y_pred_knc = modelo_knc.predict(X_val)
print('Acurácia de treino: ', modelo_knc.score(X_treino, y_treino))
print('Acurácia de validação: ', modelo_knc.score(X_val, y_val))
print('Acurácia de teste: ', modelo_knc.score(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

modelo_lr = LogisticRegression(multi_class='multinomial', random_state=0, max_iter=10000)
modelo_lr.fit(X_treino, y_treino)
y_pred_lr = modelo_lr.predict(X_val)    
print('Acurácia de treino: ', modelo_lr.score(X_treino, y_treino))
print('Acurácia de validação: ', modelo_lr.score(X_val, y_val))
print('Acurácia de teste: ', modelo_lr.score(X_test, y_test))
```

```python
from sklearn.linear_model import RidgeClassifier 

modelo_rc = RidgeClassifier(random_state=0)
modelo_rc.fit(X_treino, y_treino)
y_pred_rc = modelo_rc.predict(X_val)
print('Acurácia de treino: ', modelo_rc.score(X_treino, y_treino))
print('Acurácia de validação: ', modelo_rc.score(X_val, y_val))
print('Acurácia de teste: ', modelo_rc.score(X_test, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gb = GradientBoostingClassifier(random_state=0)
model_gb.fit(X_treino, y_treino)
y_pred_gb = model_gb.predict(X_val)
print('Acurácia de treino: ', model_gb.score(X_treino, y_treino))
print('Acurácia de validação: ', model_gb.score(X_val, y_val))
print('Acurácia de teste: ', model_gb.score(X_test, y_test))
```

```python
from matplotlib import pyplot as plt

model_names = ['Random Forest', 'Decision Tree', 'Extra Tree', 'KNeighbors', 'Logistic Regression', 'Ridge Classifier', 'Gradient Boosting']
model_scores = [modelo_rf.score(X_test, y_test), modelo_dt.score(X_test, y_test), modelo_et.score(X_test, y_test), modelo_knc.score(X_test, y_test), modelo_lr.score(X_test, y_test), modelo_rc.score(X_test, y_test), model_gb.score(X_test, y_test)]

plt.plot(model_names, model_scores, marker='o')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x em 45 graus e alinha à direita
plt.ylabel('Score')
plt.title('Comparação de Modelos')
plt.grid(True)

plt.show()
```

```python
print(classification_report(y_val, y_pred_rf_balanced))
ConfusionMatrixDisplay.from_predictions(y_val, y_pred_rf_balanced, normalize='true', cmap='Blues', display_labels=modelo_rf_balanced.classes_);
```

```python
from imblearn.pipeline import Pipeline as imbpipeline
```

```python
modelo = RandomForestClassifier(random_state=0, max_depth=10)
pipeline = imbpipeline([('oversample', SMOTE()), ('random_forest', modelo)])
```

```python
from sklearn.model_selection import cross_validate, StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
cv_results = cross_validate(pipeline, X, y, cv=skf, scoring='recall_weighted', return_train_score=True)
cv_results['test_score']
```

```python
mean = cv_results['test_score'].mean()
standard_deviation = cv_results['test_score'].std()

print(f'Confidence Interval: [{mean - 2*standard_deviation}, {min(mean + 2*standard_deviation, 1.0)}]')
```

```python
oversample = SMOTE(random_state=0)
X_balanced, y_balanced = oversample.fit_resample(X, y)
final_model = RandomForestClassifier(random_state=0, max_depth=10)
final_model.fit(X_balanced, y_balanced)
y_pred = final_model.predict(X_test)
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, normalize='true', cmap='Blues', display_labels=final_model.classes_);
```
